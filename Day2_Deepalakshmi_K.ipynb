## Lists vs. Tuples

Both lists and tuples are used to store collections of items in Python, but they have a key difference:

*   **Lists** are **mutable**, meaning you can change their contents after they are created. You can add, remove, or modify elements in a list.
*   **Tuples** are **immutable**, meaning you cannot change their contents after they are created. Once a tuple is defined, its elements are fixed.

**Mutability** refers to whether an object can be changed after it is created. Mutable objects can be modified in place, while immutable objects cannot.

Here's a summary:

| Feature     | Lists             | Tuples            |
|-------------|-------------------|-------------------|
| Mutability  | Mutable (changeable)| Immutable (unchangeable)|
| Syntax      | `[ ]`             | `( )`             |
| Use Cases   | Collections that need to be modified (e.g., adding items to a shopping cart) | Collections that should not be changed (e.g., coordinates, configuration settings) |

## When to use Sets

Sets are used in Python for collections of unique, unordered items. Here are some common use cases for sets:

*   **Removing duplicates:** Because sets only store unique elements, you can easily remove duplicate items from a list or other iterable by converting it to a set.
*   **Membership testing:** Checking if an element is present in a set is very fast compared to lists or tuples.
*   **Set operations:** Sets support mathematical set operations like union, intersection, difference, and symmetric difference, which are useful for comparing collections of data.
*   **Storing unique items:** When you need to store a collection of items where each item must be unique, a set is the appropriate data structure.

```python
# --- SIMPLE CALCULATOR ---
print("\n--- SIMPLE CALCULATOR ---")

while True:
    print("\nEnter operation (+, -, *, /) or 'x' to exit:")
    operation = input()

    if operation == 'x':
        break

    try:
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))
    except ValueError:
        print("Invalid input. Please enter valid numbers.")
        continue # Continue to the next iteration of the while loop

    if operation == '+':
        result = num1 + num2
        print(f"Result: {result}")
    elif operation == '-':
        result = num1 - num2
        print(f"Result: {result}")
    elif operation == '*':
        result = num1 * num2
        print(f"Result: {result}")
    elif operation == '/':
        if num2 != 0:
            result = num1 / num2
            print(f"Result: {result}")
        else:
            print("Error: Cannot divide by zero")
    else:
        print("Invalid operation")

print("Calculator closed.")
```

# EXERCISE 1: Number Guessing Game

Guess a number between 1 and 10. You have 5 attempts.  
Enter your guess: 5  
Too low!  
Attempts remaining: 4  
Enter your guess: 6  
Too low!  
Attempts remaining: 3  
Enter your guess: 8  
üéâ Correct! You won in 3 attempts!  

```python
# Exercise 1 - Number Guessing Game

import random # To get a random number

# Initialise variables
min_value = 1
max_value = 10
max_attempts = 5
# Get a random number within the specified range
random_number = random.randint(min_value, max_value)
# Display text
print(f"Guess a number between {min_value} and {max_value}. You have {max_attempts} attempts")
# Counter
i = 0
while i < max_attempts:

  try:
    # Get user input
    user_input = int(input("\nEnter your guess: "))
    # if user has entered the correct value
    if user_input == random_number:
      success = True
      break
    else:
      # if incorrect value entered
      success = False
      # Based on the value, show 'too low' or 'too high'
      if user_input < random_number:
        print("Too low!")
      else:
        print("Too high!")
  # if user input is errorneous
  except ValueError:
    success = False
    print(f"Error input")

  # Increment counter (if error input / incorrect input)
  i += 1
  # Show remaining attempts
  if i < max_attempts:
    print(f"Attempts remaining:", max_attempts-i, sep=" ")

# Show output text based on counter value
if i == 0:
  output_txt = "attempt"
else:
  output_txt = "attempts"
# if correct value entered, show success message
if success:
  print(f"Correct! You won in {i+1} {output_txt}!")
else:
  print(f"You ran out of attempts. The correct number was {random_number}")
```

# EXERCISE 2: List Filtering

EXERCISE 2: List Filtering  
Original list: [12, 7, 25, 3, 18, 9, 32, 15, 6]  
Even numbers: [12, 18, 32, 6]  
Numbers > 10: [12, 25, 18, 32, 15]  

```python
# Exercise 2: List Filtering

original_list = [12, 7, 25, 3, 18, 9, 32, 15, 6]  # Original list
even_numbers = []                                 # even_numbers list
numbers_greater_than_10 = []                      # numbers > 10 list

# for all values in original list
for num1 in original_list:
  # if even number
  if (num1 % 2) == 0:
    even_numbers.append(num1)
  # if greater than 10
  if num1 > 10:
    numbers_greater_than_10.append(num1)

# Print lists
print(f"Original list: {original_list}")
print(f"Even numbers: {even_numbers}")
print(f"Numbers > 10: {numbers_greater_than_10}")

```
