# Assignment 1 - Simple Calculator

```python
# Assignment 1 - Simple Calculator

# Input first number
first_number_string = input("Enter first number: ")
first_number = int(first_number_string)

# Input second number
second_number_string = input("Enter second number: ")
second_number = int(second_number_string)

# Input operation
selected_operation = input ("Enter operation (+,-,*,/): ")

# Output value
if selected_operation == "+":
  # Addition
  print (f"{first_number} + {second_number} = {first_number + second_number}")
elif selected_operation == "-":
  # Subtraction
  print (f"{first_number} - {second_number} = {first_number - second_number}")
elif selected_operation == "*":
  # Multiplication
  print (f"{first_number} * {second_number} = {first_number * second_number}")
elif selected_operation == "/":
  # Division
  print (f"{first_number} / {second_number} = {first_number / second_number}")
else:
  # Any other value
  print (f"Wrong operation entered")

# Thank you message
print ("Thank you!", end="\n")
```

# Assignment 2 - Temperature Converter

```python
# Assignment 2 - Temperature Converter

# Input temperature
temperature_string = input("Enter temperature: ")
temperature = int(temperature_string)

# Input temperature scale
scale_string = input("(C)elsius or (F)ahrenheit? ")

# Output value
if scale_string == "C":
  # Convert to Fahrenheit
  print (f"{temperature} C = {(temperature * 9/5) + 32} F")
elif scale_string == "F":
  # Convert to Celsius
  print (f"{temperature} F = {(temperature - 32) * 5/9} C")
else:
  print (f"Wrong value entered", end="\n")
```
